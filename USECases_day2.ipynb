# Working with Dbutils & Widgets

#####Defining the Paths

```python
base_path = "/Volumes/workspace/default/volumewd47"
#Create raw_path, processed_path and archive_path as given below...
raw_path = f"{base_path}/raw"
processed_path = f"{base_path}/processed"
archive_path = f"{base_path}/archive"
print("RAW ----->", raw_path)
print("PROCESSED ----->", processed_path)
print("ARCHIVE ----->", archive_path)
```

#####Creating the volume

```python
%sql
create volume if not exists default.volumewd47;
```

#####Creating following directories using Dbutils

```python
dbutils.fs.mkdirs(raw_path)
dbutils.fs.mkdirs(processed_path)
dbutils.fs.mkdirs(archive_path)
```

#####Creating Sample Data file and writing it into the raw folder

```python
sample_data = """patient_id,patient_name,age,gender\n1,John Doe,68,M 2,Jane Smith,54,F """
dbutils.fs.put(raw_path+"/sample_data.csv", sample_data, True)
df = spark.read.csv(raw_path+"/sample_data.csv", header=True)
display(df)
```

#####Listing files under Raw folder

```python
dbutils.fs.ls(raw_path)
```

#####Copying files from Raw to Processed folder

```python
dbutils.fs.cp(raw_path+"/", processed_path+"/", True)
dbutils.fs.ls(processed_path)
```

#####Creating widgets

```python
dbutils.widgets.text("Owner Name :","Kunal")
owner_name_wdg = dbutils.widgets.get("Owner Name :")
dbutils.widgets.dropdown("Environment :","dev",["dev","qa","prod"])
environment_wdg = dbutils.widgets.get("Environment :")
```

```python
print("Owner --->",owner_name_wdg)
print("Environment --->",environment_wdg)
```

#####Moving the processed file to archive

```python
dbutils.fs.mv(processed_path+"/", archive_path+"/",True)
dbutils.fs.ls(archive_path)
```
