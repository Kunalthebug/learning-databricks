### Fetching the Data from API using get request

```python
import requests

headers = {"Accept": "text/csv"}
uri_path = "https://public.tableau.com/app/sample-data/mobile_os_usage.csv"
vol_path = "/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv"
data = requests.get(uri_path,headers=headers)
data = data.content.decode("utf-8")
```

### Writing the data into the volume **usage_metrics**

```python
with open(vol_path, mode="w", newline="", encoding="utf-8") as file:
    file.write(data)

print(f"Successfully wrote data to {vol_path}")
```

### Using dbutils feature

```python
dbutils.fs.put("/Volumes/workspace/default/usage_metrics/mobile_os_usage.csv",data,overwrite=True)
```
